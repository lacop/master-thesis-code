```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as PD
from collections import defaultdict
```

```python
data = np.genfromtxt('../experiments/ex1-sha3-minisat-bos.csv',
                     delimiter=',', dtype=None,
                     #dtype=[('msglen', '<i8'), ('rounds', 'f8'), ('outrefcnt', '<i8'), ('solver', 'S28'), ('bo', 'S11'), ('seed', '<i8'), ('confl', '<i8'), ('time', '<f8')],
                     names=['msglen', 'rounds', 'outrefcnt', 'solver',
                            'bo', 'seed', 'confl', 'time'])
```

```python
f = data['bo']==b'none'
plt.scatter(data[f]['rounds'], data[f]['time'], c=[(1,0,0)]*len(f))
f = data['bo']==b'rlast-S-x-y'
plt.scatter(data[f]['rounds'], data[f]['time'], c=[(0,1,0)]*len(f), alpha=0.25)
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
def compare(ax, filt):
    ax.set_yscale('log')
    ax.set_title(filt.decode())
    f = data['bo']==b'none'
    sets = defaultdict(list)
    for r, t in data[f][['rounds', 'time']]:
        sets[r].append(t)
    xpos = sorted(sets.keys())
    viols = [sets[x] for x in xpos]
    xpos = [x for x in xpos]
    _ = ax.violinplot(viols, positions=xpos, showmeans=True, showmedians=True, widths=0.25)

    f = data['bo']==filt
    sets = defaultdict(list)
    for r, t in data[f][['rounds', 'time']]:
        sets[r].append(t)
    xpos = sorted(sets.keys())
    viols = [sets[x] for x in xpos]
    xpos = [x+0.25 for x in xpos]
    parts = ax.violinplot(viols, positions=xpos, showmeans=True, showmedians=True, widths=0.25)
    for part in parts['bodies']:
        part.set_facecolor('green')

compare(fig.add_subplot(2,2,1), b'r0-S-x-y')
compare(fig.add_subplot(2,2,2), b'r0-S-y-x')
compare(fig.add_subplot(2,2,3), b'rlast-S-x-y')
compare(fig.add_subplot(2,2,4), b'rlast-S-y-x')

plt.savefig('violin-compare-bos.pdf')
        
#plt.violinplot(list(data[f][['rounds','time']]))
#plt.violinplot([[1, 10, 20, 30, 11, 13, 12], [1, 15], [2, 20]], showmedians=True)
```
