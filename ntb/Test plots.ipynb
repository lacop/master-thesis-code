```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as PD
from collections import defaultdict
```

```python
data = np.genfromtxt('../experiments/ex1-sha3-minisat-bos.csv',
                     delimiter=',', dtype=None,
                     #dtype=[('msglen', '<i8'), ('rounds', 'f8'), ('outrefcnt', '<i8'), ('solver', 'S28'), ('bo', 'S11'), ('seed', '<i8'), ('confl', '<i8'), ('time', '<f8')],
                     names=['msglen', 'rounds', 'outrefcnt', 'solver',
                            'bo', 'seed', 'confl', 'time'])
```

```python
f = data['bo']==b'none'
plt.scatter(data[f]['rounds'], data[f]['time'], c=[(1,0,0)]*len(f))
f = data['bo']==b'rlast-S-x-y'
plt.scatter(data[f]['rounds'], data[f]['time'], c=[(0,1,0)]*len(f), alpha=0.25)
plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
def compare(ax, filt):
    #ax.set_yscale('log')
    ax.set_title('none vs. ' + filt.decode())
    f = data['bo']==b'none'
    sets = defaultdict(list)
    for r, t in data[f][['rounds', 'time']]:
        sets[r].append(t)
    xpos = sorted(sets.keys())
    viols = [sets[x] for x in xpos]
    xpos = [x-0.0125-0.075 for x in xpos]
    parts = ax.violinplot(viols, positions=xpos, showmeans=True, showmedians=True, widths=0.25)
    parts['cbars'].set_linewidth(0.25)
    parts['cmaxes'].set_linewidth(0.5)
    parts['cmins'].set_linewidth(0.5)
    parts['cmeans'].set_linewidth(0.5)
    parts['cmedians'].set_linewidth(0.5)
    parts['cmedians'].set_dashes('--')

    f = data['bo']==filt
    sets = defaultdict(list)
    for r, t in data[f][['rounds', 'time']]:
        sets[r].append(t)
    xpos = sorted(sets.keys())
    viols = [sets[x] for x in xpos]
    xpos = [x+0.125+0.075 for x in xpos]
    parts = ax.violinplot(viols, positions=xpos, showmeans=True, showmedians=True, widths=0.25)
    parts['cbars'].set_linewidth(0.25)
    parts['cmaxes'].set_linewidth(0.5)
    parts['cmins'].set_linewidth(0.5)
    parts['cmeans'].set_linewidth(0.5)
    parts['cmedians'].set_linewidth(0.5)
    parts['cmedians'].set_dashes('--')
    for part in parts['bodies']:
        part.set_facecolor('green')
    ax.set_xlabel('Rounds')
    ax.set_ylabel('Time')
    #return parts

#compare(fig.add_subplot(2,2,1), b'r0-S-x-y')
#compare(fig.add_subplot(2,2,2), b'r0-S-y-x')
#compare(fig.add_subplot(2,2,3), b'rlast-S-x-y')
#compare(fig.add_subplot(2,2,4), b'rlast-S-y-x')

compare(fig.add_subplot(1,1,1), b'r0-S-x-y')

#plt.savefig('violin-compare-bos.pdf')
#plt.savefig('violin-compare-bos.png')
        
#plt.violinplot(list(data[f][['rounds','time']]))
#plt.violinplot([[1, 10, 20, 30, 11, 13, 12], [1, 15], [2, 20]], showmedians=True)
```

```python
k1 = b'none'
k2 = b'r0-S-x-y'
diffs = defaultdict(list)
for k1row in data[data['bo'] == k1]:
    k2row = data[(data['bo'] == k2) & (data['rounds'] == k1row['rounds']) & (data['seed'] == k1row['seed'])][0]
    diffs[k1row['rounds']].append(k1row['time']-k2row['time'])
#diffs
#data
xpos = sorted(diffs.keys())
diffs = [diffs[x] for x in xpos]

fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Rounds')
ax.set_ylabel('Time difference')
ax.set_title(k1.decode() + ' vs ' + k2.decode())
_ = ax.violinplot(diffs, positions=xpos, showmeans=True, showmedians=True, widths=0.5)
ax.axhline(y=0, color='k')

#plt.savefig('violin-compare-bos-diff.png')
```
