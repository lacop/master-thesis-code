```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
#import pandas as PD
from collections import defaultdict
```

```python
data = np.genfromtxt('../experiments/ex1-sha3-minisat-bos.csv',
#data = np.genfromtxt('../experiments/ex2-sha3-minisat-bos.csv',
                     delimiter=',', dtype=None,
                     #dtype=[('msglen', '<i8'), ('rounds', 'f8'), ('outrefcnt', '<i8'), ('solver', 'S28'), ('bo', 'S11'), ('seed', '<i8'), ('confl', '<i8'), ('time', '<f8')],
                     names=['msglen', 'rounds', 'outrefcnt', 'solver',
                            'bo', 'seed', 'confl', 'time'])
```

```python
best_times = defaultdict(lambda: defaultdict(lambda: defaultdict(lambda: float('inf')))) # best[BO][rounds][seed] = best time
best_confl = defaultdict(lambda: defaultdict(lambda: defaultdict(lambda: float('inf')))) # best[BO][rounds][seed] = conflicts
for row in data:
    # Timeouts
    time = row['time']
    if time == b'?':
        # Skip?
        #continue|
        time = 60 + row['seed'] # Avoid singular matrix
        row['confl'] = 10000 + row['seed']
    if type(time) == np.bytes_:
        time = float(time.decode())
    best_times[row['bo']][row['rounds']][row['seed']] = min(best_times[row['bo']][row['rounds']][row['seed']], time)
    if row['rounds'] > 1: # With one round all instances have 0 conflicts, skip
        best_confl[row['bo']][row['rounds']][row['seed']] = min(best_confl[row['bo']][row['rounds']][row['seed']], row['confl'])
```

```python
def violin(ax, xpos, ylist, offset=0, color='yellow'):
    xpos = [x+offset for x in xpos]
    parts = ax.violinplot(ylist, positions=xpos, showmeans=True, showmedians=True, widths=0.2)
    parts['cbars'].set_linewidth(0.05)
    parts['cmaxes'].set_linewidth(0.1)
    parts['cmins'].set_linewidth(0.1)
    parts['cmeans'].set_linewidth(0.1)
    parts['cmedians'].set_linewidth(0.1)
    parts['cmedians'].set_dashes('--')
    for part in parts['bodies']:
        part.set_facecolor(color)
        part.set_linewidth(0.01)

def plot_bo_best(arr, ax, bo, offset=0, color='yellow'):
    xpos = sorted(arr[bo].keys())
    violin(ax, xpos, [list(arr[bo][x].values()) for x in xpos], offset, color)
```

```python
def plot_side_by_side(arr, label, ax, bo1, bo2):
    ax.set_title(bo1.decode() + ' [yellow] vs. ' + bo2.decode() + ' [green]')
    ax.set_xlabel('Rounds')
    ax.set_ylabel(label)
    #ax.set_yscale('log')
    
    plot_bo_best(arr, ax, bo1, -0.175)
    plot_bo_best(arr, ax, bo2, 0.175, 'green')
    
def plot_differences(arr, label, ax, bo1, bo2):
    ax.set_title(bo1.decode() + ' [A] vs. ' + bo2.decode() + ' [B]')
    ax.set_xlabel('Rounds')
    ax.set_ylabel(label + ' difference (B-A)')
    ax.annotate('[A] is better', (0.1, 0.95), xycoords='axes fraction')
    ax.annotate('[B] is better', (0.1, 0.05), xycoords='axes fraction')
    ax.axhline(y=0, color='k')
    
    assert (arr[bo1].keys() == arr[bo2].keys())
    xpos = sorted(arr[bo1].keys())
    diffs = [[arr[bo2][x][s] - arr[bo1][x][s] for s in arr[bo1][x].keys()] for x in xpos]
    violin(ax, xpos, diffs)
    
def plot_ratios(arr, label, ax, bo1, bo2):
    ax.set_title(bo1.decode() + ' [A] vs. ' + bo2.decode() + ' [B]')
    ax.set_xlabel('Rounds')
    ax.set_yscale('log')
    ax.set_ylabel(label + ' ratio (B/A)')
    ax.annotate('[A] is better', (0.1, 0.90), xycoords='axes fraction')
    ax.annotate('[B] is better', (0.1, 0.1), xycoords='axes fraction')
    ax.axhline(y=1, color='k')
    
    assert (arr[bo1].keys() == arr[bo2].keys())
    xpos = sorted(arr[bo1].keys())
    ratios = [[arr[bo2][x][s] / arr[bo1][x][s] for s in arr[bo1][x].keys()] for x in xpos]
    for i in range(len(ratios)): # Hack to avoid singular matrix if it makes no difference at all
        if all(ratios[i][x] == 1.0 for x in range(len(ratios[i]))):
            ratios[i].append(1.1)
            ratios[i].append(0.9)
    violin(ax, xpos, ratios)
```

```python
def compare_bos(bo1, bo2):
    fig = plt.figure(figsize=(20,10))
    fig.suptitle(bo1.decode() + ' vs. ' + bo2.decode(), fontsize=16)
    plot_side_by_side(best_times, 'Time', fig.add_subplot(2,2,1), bo1, bo2)
    plot_side_by_side(best_confl, 'Conflicts', fig.add_subplot(2,2,2), bo1, bo2)
    #plot_differences(best_times, 'Time', fig.add_subplot(2,2,3), bo1, bo2)
    plot_ratios(best_times, 'Time', fig.add_subplot(2,2,3), bo1, bo2)
    #plot_differences(best_confl, 'Conflicts', fig.add_subplot(2,2,4), bo1, bo2)
    plot_ratios(best_confl, 'Conflicts', fig.add_subplot(2,2,4), bo1, bo2)
```

```python
compare_bos(b'none', b'r0-S-x-y')
compare_bos(b'none', b'r0-S-y-x')
```
